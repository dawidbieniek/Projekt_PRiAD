# Biblioteki

```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [12, 6]
```

# Wczytanie danych

```python
stars_raw = pd.read_csv('stars.csv')
stars_raw.head()
```

# Podstawowe informacje o danych

```python
print(stars_raw.info())
print('\n', stars_raw.describe())
print('\nLiczba pustych komórek')
print(stars_raw.isnull().sum())
```

```python
sns.pairplot(stars_raw, hue='Star type', palette='Dark2')
sns.pairplot(stars_raw)
```

```python
sns.histplot(stars_raw['Temperature (K)'])
plt.show()
```

```python
sns.histplot(stars_raw['Luminosity(L/Lo)'])
plt.show()
```

```python
sns.histplot(stars_raw['Radius(R/Ro)'])
plt.show()
```

```python
sns.histplot(stars_raw['Absolute magnitude(Mv)'])
plt.show()
```

```python
#sns.histplot(stars_raw['Star type'])
stars_raw['Star type'].value_counts().plot(kind='bar')# taki wykres lepiej rozmieszcza wartości
plt.show()
print('Typy gwiazd:', stars_raw['Star type'].unique())
print('Liczba typów gwiazd:', len(stars_raw['Star type'].unique()))
```

```python
plt.xticks(rotation=90)
sns.histplot(stars_raw['Star color'])
plt.show()
colors = stars_raw['Star color'].unique()
print('Kolory gwiazd (', len(colors), '):')
for c in colors:
    print('\t|' + c + '|')
```

```python
print(stars_raw.loc[stars_raw['Star color'] == 'Blue white '])
print(stars_raw.loc[stars_raw['Star color'] == 'Blue white'])
print(stars_raw.loc[stars_raw['Star color'] == 'Blue-White'])

#stars_raw.loc[stars_raw['Star color'] == 'Blue-White'].corrwith(stars_raw.loc[stars_raw['Star color'] == 'Blue white '])
```

```python
sns.histplot(stars_raw['Spectral Class'])
plt.show()
print('Klasy spektralne:', stars_raw['Spectral Class'].unique())
print('Liczba klas spektralnych:', len(stars_raw['Spectral Class'].unique()))
```

```python
print('Rozmieszczenie gwiazd w zależności od temperatury i wiekości')
sns.lmplot(x='Temperature (K)', y='Absolute magnitude(Mv)', data=stars_raw, hue='Star type', legend=False, fit_reg=False)
```

```python
stars_cat = stars_raw['Star type']
stars_attr = stars_raw.drop(['Star type'], axis=1)

print(stars_cat)
stars_attr.info()
```

```python
sns.heatmap(stars_attr.corr(), annot=True)
```

```python
sns.boxplot(data=stars_attr)
```

# Skalowanie

```python
scaler = StandardScaler()
stars_attr.iloc[:, :4] = scaler.fit_transform(stars_attr.iloc[:, :4])
sns.boxplot(data=stars_attr)
```

# Atrybuty kategoryczne

```python
encoder = OrdinalEncoder()
stars_attr[['Star color', 'Spectral Class']] = encoder.fit_transform(stars_attr[['Star color', 'Spectral Class']])
stars_attr.info()
sns.heatmap(stars_attr.corr(), annot=True)
```
